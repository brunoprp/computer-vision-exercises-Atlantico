# Tests detection model with images

```python
from objetc_detect import Detect
from PIL import Image, ImageDraw, ImageFont
from IPython.display import clear_output
import matplotlib.pyplot as plt
```

```python
# Load model
checkpoint = 'checkpoint_ssd300.pth.tar'
obj_detect = Detect()
model = obj_detect.loadModel(checkpoint)
```

```python
#Load imgade and detect objetct with model
img_path = 'image-test/img2.jpg' # image path
original_image = Image.open(img_path, mode='r')
original_image = original_image.convert('RGB')
img_detect = obj_detect.detectImg(original_image, model)
clear_output()
```

```python
# Plot image object detect 
plt.figure(figsize=(15,9))
plt.axis('off')
plt.imshow(img_detect)
```
